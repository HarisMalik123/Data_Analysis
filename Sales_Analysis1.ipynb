```python
##importing pandas 
import pandas as pd
```

```python
#Reading Sales.Csv File
pd.read_csv('sales.csv')
```

```python
##storing it in variable
data=pd.read_csv('sales.csv')
```

```python
data
```

```python
## this expression will return a new DataFrame containing only the rows where the 'Gender' column has the value 'Male'.
data.query('Gender=="Male"')
```

```python
## data.sum() this calculate the sum of each column in our data frame and it will return series where each index represent a particular columns
data.sum()
```

```python
##If we Mention columns name it will return sum of all values in that particular column
data.sum()['Total']
```

```python
data.max()
```

```python
 ##it will return maximum values in each column but if we mention data.max()['ColumnName'] it will return maximum from it
data.max()
```

```python
data.max()['Total']
```

```python
## it is used to filter a DataFrame named data to include only the rows where the value in the 'Total' column is equal to the maximum value in the entire 'Total' column.
## it will create a bolean mask 
data[data['Total']==data.max()['Total']]
```

```python
##This groups the DataFrame data by the unique values in the 'City' column. It doesn't perform any computation on its own it is used for further operation
data.groupby('City')
```

```python
##first it will group the data by unique values in city then it will sum all values in them
data.groupby('City').sum()
```

```python
##After performing sum() then we searched for particular column Named 'Total'
data.groupby('City').sum()['Total']
```

```python
##We Group the data with respect to member column
data.groupby('Member').sum()
```

```python
##We searched a total number of members who said yes or no
data.groupby('Member').sum()['Total']
```
